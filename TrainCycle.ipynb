```python
import cv2
from darkflow.net.build import TFNet
import matplotlib.pyplot as plt
from sys import exit

%config InlineBackend.figure_format = 'svg'
```

```python
# define the model options and run

options = {
    'model': 'cfg/tiny-yolo-voc-3c.cfg',
    'load': 'bin/tiny-yolo-voc.weights',
    'threshold': 0.3,
}

tfnet = TFNet(options)
```

```python
options = {"model": "cfg/tiny-yolo-voc-3c.cfg", 
           "load": "bin/tiny-yolo-voc.weights",
           "batch": 8,
           "epoch": 100,
           "train": True,
           "annotation": "train/annotations/",
           "dataset": "train/images/"}
```

```python
tfnet = TFNet(options)
```

```python
tfnet.train()
```

```python
img = cv2.imread('test2.jpeg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# use YOLO to predict the image
result = tfnet.return_predict(img)
print(result)

img.shape
```

```python
tl = (result[1]['topleft']['x'], result[1]['topleft']['y'])
br = (result[1]['bottomright']['x'], result[1]['bottomright']['y'])
label = result[1]['label']


# add the box and label and display it
img = cv2.rectangle(img, tl, br, (0, 255, 0), 7)
img = cv2.putText(img, label, tl, cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 0), 2)
plt.imshow(img)
plt.show()
```
